```python
%matplotlib inline
import ystockquote
import numpy as np
from collections import namedtuple
import matplotlib.pyplot as plt

quotes = ystockquote.get_historical_prices('aapl', '2013-01-03', '2017-01-08')
quotes['2013-12-26'].keys()
```

```python
from pandas import Series, Timestamp
```

```python
dates = [Timestamp(t) for t in quotes.keys()]
```

```python
adj_points = sorted([(v['Adj Close'], k) for k, v in quotes.items()], key=lambda x: x[1])
points = sorted([(v['Close'], k) for k, v in quotes.items()], key=lambda x: x[1])
```

```python
adj_values, dates = zip(*adj_points)
values, dates = zip(*points)
adj_ts = Series(adj_values, [Timestamp(d) for d in dates])
ts = Series(values, [Timestamp(d) for d in dates])
```

```python
plt.figure()
plt.plot(adj_ts)
```

```python
plt.figure()
plt.plot(ts)
```
